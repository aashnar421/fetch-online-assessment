```python
#importing relevant libraries
import json
import datetime
import numpy as np
```

```python
#function to open json files
def open_json(file, var_name):
    var_name = []
    for line in open(file, 'r'):
        var_name.append(json.loads(line))
    return var_name
```

#### Checking see if there are any duplicate entries in the 3 tables:

```python
#getting unique entries.
def num_unique_users(data):
    unique_users = []
    for entry in data:
        user_id = entry['_id']['$oid']
        if user_id not in unique_users:
            unique_users.append(user_id)

    print('total entries in database: ' + str(len(data)))
    print('total unique entries: ' + str(len(unique_users)))
#list, dictionary, dictionary
```

```python
num_unique_users(users)
```

This shows that there is a lot of duplicate data in the users database. Either each entry is duplicated, or there's a possibility that for each user, we have conflicting info (e.g. they might have 5 different states they live in, according to the data). We need to see if there is simply duplicate data, which is easy to remove, or if there's conflicting data.

```python
for entry in users:
    if entry['_id']['$oid'] == '54943462e4b07e684157a532':
        print(entry)

#manual checking shows that all entries are duplicates of each other. state is missing.
```

### Digging deeper into the receipts table:

Diving deeper into the 'rewardsReceiptItemList' in each receipt entry. I noticed while looking through the data that some of the rewardsReceiptItemList items were duplicated in the same receipt. Checking if this is the case in one such receipt:

```python
receiptid = "5f9c74f70a7214ad07000037"
for receipt in receipts:
    if receipt['_id']['$oid'] == receiptid:
        receipt_items_entry = receipt['rewardsReceiptItemList']
        for k in receipt_items_entry:
            item_entries = []
            if 'userFlaggedBarcode' in k:
                item_entries.append(k['userFlaggedBarcode'])
            if 'userFlaggedDescription' in k:
                item_entries.append(k['userFlaggedDescription'])
            if 'userFlaggedPrice' in k:
                item_entries.append(k['userFlaggedPrice'])
            if 'userFlaggedQuantity' in k:
                item_entries.append(k['userFlaggedQuantity'])
            if 'partnerItemId' in k:
                item_entries.append(k['partnerItemId'])
            print(item_entries)
```

This shows that apart from the first item that the customer bought, the rest are all the same item (seem like, upon first glance). They all have the same barcode, description, and price. The only difference is the 'partnerItemId'. There are 2 issues with this:

1. This field is ambiguous - who is the partner in this case? The customer, the brand, or a third party? Does it refer to the brand's internal catalogue of the item? And if so, then why does the brand have 10 different uuids for the item? If this is the case, then we should flag it to the brand.
2. There is a 'quantity' field - why is this item listed 10 individual times, each time saying only 1 can was purchased, when instead it could've been one singular entry, with 'quantitypurchased' = 10? Is it to do with the PartnerItemId?

### Checking date validity in the users data

```python
def convert_to_datetime(timestamp):
    return datetime.datetime.fromtimestamp(timestamp/1000)
```

```python
invalid_entries = 0
for user in users:
    createdDateTs = user['createdDate']['$date']
    print("Created at: ", createdDateTs)
    if 'lastLogin' in user:        
        lastLoginDateTs = user['lastLogin']['$date']
        timediff = lastLoginDateTs - createdDateTs
        if timediff <= 0:
            invalid_entries += 1
        print("lastLogin: ",lastLoginDateTs)
        print('timediff: ', lastLoginDateTs - createdDateTs)
    else:
        print("No last login")
    print("\n")
```

Looks like these two fields are alright.
